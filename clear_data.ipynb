```python
import json
import os
import sys
import random

import json
import random

def process_jsonl_files(input_files, output_train, output_dev, output_test, random_seed=42):
    merged_data = []
    
    for file_path in input_files:
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                for line in f:
                    if line.strip():  
                        merged_data.append(json.loads(line.strip()))
        except FileNotFoundError:
            print(f"Lỗi: Không tìm thấy file {file_path}")
            return

    total_records = len(merged_data)
    if total_records == 0:
        print("Không có dữ liệu trong các file đầu vào.")
        return

    random.seed(random_seed) 
    random.shuffle(merged_data)

    train_end = int(total_records * 0.7)
    dev_end = train_end + int(total_records * 0.1)

    train_data = merged_data[:train_end]
    dev_data = merged_data[train_end:dev_end]
    test_data = merged_data[dev_end:] 
    def write_jsonl(data, filename):
        with open(filename, 'w', encoding='utf-8') as f:
            for item in data:
                f.write(json.dumps(item, ensure_ascii=False) + '\n')

    write_jsonl(train_data, output_train)
    write_jsonl(dev_data, output_dev)
    write_jsonl(test_data, output_test)

    print(f"Tổng số bản ghi: {total_records}")
    print(f"Đã chia thành công!")
    print(f"- Train (70%): {len(train_data)} bản ghi -> {output_train}")
    print(f"- Dev   (10%): {len(dev_data)} bản ghi -> {output_dev}")
    print(f"- Test  (20%): {len(test_data)} bản ghi -> {output_test}")

file1 = "D:\\Python\\ABSA_targetted\\Data\\hotel.jsonl"
file2 = "D:\\Python\\ABSA_targetted\\Data\\restaurant.jsonl"
file3 = "D:\\Python\\ABSA_targetted\\Data\\mobile.jsonl"
input_files = [file1, file2, file3]
output_train = 'train.jsonl'
output_dev = 'dev.jsonl'
output_test = 'test.jsonl'

process_jsonl_files(input_files, output_train, output_dev, output_test)
```

```python
import json
import os

def clean_absa_dataset(input_filepath: str, output_filepath: str):
    valid_samples = []
    removed_count = 0

    with open(input_filepath, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            try:
                sample = json.loads(line)

                is_label_empty = not sample.get("label") 
                
                is_labels_null = "labels" in sample and sample["labels"] is None

                if is_label_empty or is_labels_null:
                    removed_count += 1
                    continue 
                valid_samples.append(sample)
                
            except json.JSONDecodeError:
                print(f"⚠️ Lỗi parse JSON ở dòng: {line}")

    with open(output_filepath, 'w', encoding='utf-8') as f:
        for sample in valid_samples:
            f.write(json.dumps(sample, ensure_ascii=False) + '\n')

    print(f"📄 Xử lý file '{input_filepath}':")
    print(f"   + Giữ lại: {len(valid_samples)} mẫu")
    print(f"   - Đã xóa : {removed_count} mẫu lỗi\n")

dataset_splits = ["D:\\Python\\ABSA_targetted\\Data\\train.jsonl", "D:\\Python\\ABSA_targetted\\Data\\dev.jsonl", "D:\\Python\\ABSA_targetted\\Data\\test.jsonl"]

for split in dataset_splits:
    if os.path.exists(split):
        output_file = f"{split}"
        clean_absa_dataset(split, output_file)
    else:
        print(f"❌ Không tìm thấy file: {split}. Hãy kiểm tra lại đường dẫn.")
```
